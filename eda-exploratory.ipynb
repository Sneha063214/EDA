```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
#so that resulant output can be obtained just after the line of code
```

```python
df=pd.read_csv('/kaggle/input/zomatodata/zomato.csv',encoding='latin-1')#latin-1:to include all types of datatypes
df.head()
```

```python
df_country=pd.read_excel('/kaggle/input/zomatodata/Country-Code (1).xlsx')
df_country.head()
```

```python
final_df=pd.merge(df,df_country,on='Country Code',how='left')
```

```python
final_df.head()
```

```python
country_names=final_df.Country.value_counts().index
```

```python
country_val=final_df.Country.value_counts().values
```

```python
plt.pie(country_val[:3],labels=country_names[:3],autopct='%1.2f%%')#piechart-top3 counties that use zomato
```

```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Ratingcount'})
```

```python
ratings
```

```python
plt.rcParams['figure.figsize']=(12,6)
sns.barplot(x="Aggregate rating",y="Ratingcount",data=ratings)
```

```python
sns.barplot(x="Aggregate rating",y="Ratingcount",hue="Rating color",data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

```python
sns.countplot(x="Rating color",data=ratings,palette=['blue','red','orange','yellow','lightgreen','green'])
```
